```python
def apply_discount(price, discount):
  ## Quando apply_discount é chamado com um price (primeiro argumento) que não é um número (int ou float) ele deve retornar The price should be a number.
    if not isinstance(price, (int, float)):
        return "The price should be a number."
    if not isinstance(discount, (int, float)):
        return "The discount should be a number."
    if price <= 0:
        return "The price should be greater than 0."
    if discount <= 0 or discount >= 100:
        return "The discount should be between 0 and 100."
    discount = 100 - discount
    final_discount = price * (discount/100)
    return final_discount

print(apply_discount(74.5, 20.0))
print(apply_discount("abc", 20.0))
```

```python
full_dot = '●'
empty_dot = '○'

def create_character(name, strenght, qi, carism):
    if type(name) != str:
        return 'The character name should be a string'
    if name == '':
        return 'The character should have a name'
    if len(name) > 10:
        return 'The character name is too long'
    if name.count(" "):
        return 'The character name should not contain spaces'

    if type(strenght) != int or type(qi) != int or type(carism) != int:
        return 'All stats should be integers'
    if strenght < 1 or qi < 1 or carism < 1:
        return 'All stats should be no less than 1'
    if strenght > 4 or qi > 4 or carism > 4:
        return 'All stats should be no more than 4'
    if (strenght+qi+carism) < 7:
        return 'The character should start with 7 points'
    if strenght + qi + carism > 7:
        return 'The character should start with 7 points'

    return('ren\nSTR ●●●●○○○○○○\nINT ●●○○○○○○○○\nCHA ●○○○○○○○○○' and f'{name}\nSTR {full_dot*strenght+empty_dot*(10-strenght)}\nINT {full_dot*qi+empty_dot*(10-qi)}\nCHA {full_dot*carism+empty_dot*(10-carism)}')


print(create_character('ren', 4, 2, 1))
```

```python
def pin_extractor(poems):
    secret_codes = []
    for poem in poems:
        secret_code = ''
        lines = poem.split('\n')
        for line_index, line in enumerate(lines):
            words = line.split()
            if len(words) > line_index:
                secret_code += str(len(words[line_index]))
            else:
                secret_code += '0'
        secret_codes.append(secret_code)
    return secret_codes


poem = """Stars and the moon
shine in the sky
white and
until the end of the night"""

poem2 = 'The grass is green\nhere and there\nhoping for rain\nbefore it turns yellow'
poem3 = 'There\nonce\nwas\na\ndragon'

print(pin_extractor([poem, poem2, poem3]))
```

```python
def comprimir(texto):
    if not texto: return ""

    resultado = ""
    contagem = 1

    for i in range(1, len(texto)):
        if texto[i] == texto[i-1]:
            contagem += 1
        else:
            # Regra aplicada aqui: se contagem for 1, concatena vazio ""
            resultado += texto[i-1] + (str(contagem) if contagem > 1 else "")
            contagem = 1

    # Regra aplicada aqui também para o último caractere
    resultado += texto[-1] + (str(contagem) if contagem > 1 else "")

    return resultado

print(comprimir("aaabbc"))
```

```python
def add_setting(dictionary, tupla):
    key = tupla[0].lower()
    value = tupla[1].lower()
    if key in dictionary:
        return f"Setting '{key}' already exists! Cannot add a new setting with this name."
    if key not in dictionary:
        dictionary[key] = value
        return f"Setting '{key}' added with value '{value}' successfully!"


def update_setting(dictionary, tupla):
    key = tupla[0].lower()
    value = tupla[1].lower()
    if key in dictionary:
        dictionary[key] = value
        return f"Setting '{key}' updated to '{value}' successfully!"
    if key not in dictionary:
        return f"Setting '{key}' does not exist! Cannot update a non-existing setting."


def delete_setting(dictionary,key):
    key = key.lower()
    if key in dictionary:
        del dictionary[key]
        return f"Setting '{key}' deleted successfully!"
    else:
        return "Setting not found!"


def view_settings(dictionary):
    if not dictionary:
        return "No settings available."

    out = "\nCurrent User Settings:\n"
    for key, value in dictionary.items():
        out += f"{key.capitalize()}: {value}\n"
    return out


test_settings = {
    'theme': 'dark',
    'notifications': 'enabled',
    'volume': 'high',
    'test': 'ok'
}

print(update_setting(test_settings, ('Theme', 'Verde')))
print(delete_setting(test_settings, 'Theme'))
print(add_setting(test_settings, ('Tema', 'AZUL')))
print(view_settings(test_settings))
```

```python
# Desafio 107 - Modulo Moeda.py (a parte)
def metade(preco=0, format=bool):
  metade = preco/2
  return moeda(metade)


def dobro(preco=0, format=bool):
  dobro = preco*2
  return moeda(dobro)


def aumentar(preco=0, taxa=0, format=bool):
  aumentar = (100+taxa)/100
  aumento = preco*aumentar
  return moeda(aumento)


def diminuir(preco=0, taxa=0, format=bool):
  diminuição = (100-taxa)/100
  redução = preco*diminuição
  return moeda(redução)


def moeda(preco=0,unitario = "R$"):
  return f"{unitario}{preco:.2f}".replace(".",",")


def resumo(preco=0,mais=0,menos=0):
  print("-"*25)
  print(f"{"RESUMO DO VALOR":^25}")
  print("-"*25)
  print(f"Preço analisado: {moeda(preco):^15}")
  print(f"Dobro do preço: {dobro(preco):^15}")
  print(f"Metade do preço: {metade(preco):^15}")
  print(f"{mais}% de aumento: {aumentar(preco,mais):^15}")
  print(f"{menos}% de redução: {diminuir(preco,menos):^15}")
  print("-"*25)


def leiaDinheiro(msg):
  while True:
    num = input(msg).replace(",",".").strip()
    if num.isalpha():
      print(f'ERRO: "{num}" é um preço inválido!')
    else:
      return float(num)
```

```python
# Desafio 115a - passo a passo do modularizador, primeiro os modulos:
def leiaInt(msg):
  while True:
      try:
        numero = int(input(msg))
      except KeyboardInterrupt:
        print("Usuário desistiu D:")
        return 0
      except:
        print("ERRO: Digite um valor válido.")
      else:
        return numero

def linha(tam = 42):
  return "-" * tam

def cabeçalho(txt):
  print(linha())
  print(txt.center(42))
  print(linha())

def menu(lista):
  cabeçalho("MENU PRINCIPAL")
  c = 1
  for item in lista:
    print(f"{c} - {item}")
    c += 1
  print(linha())
  opc = leiaInt('Sua opção: ')
  return opc
```

```python
def validate_isbn(isbn, length):
    if len(isbn) != length:
        print(f'ISBN-{length} code should be {length} digits long.')
        return

    # Separa os dígitos iniciais e o dígito verificador fornecido
    main_digits = isbn[0:length-1]
    given_check_digit = isbn[-1].upper() # Garante 'X' maiúsculo para ISBN-10

    main_digits_list = [int(digit) for digit in main_digits]

    # Calcula o dígito esperado com base nos dígitos iniciais
    if length == 10:
        expected_check_digit = calculate_check_digit_10(main_digits_list)
    else:
        expected_check_digit = calculate_check_digit_13(main_digits_list)

    # Compara o dígito fornecido com o calculado
    if given_check_digit == expected_check_digit:
        print('Valid ISBN Code.')
    else:
        print('Invalid ISBN Code.')

def calculate_check_digit_10(main_digits_list):
    digits_sum = 0
    # Multiplica os 9 primeiros dígitos por pesos de 10 a 2
    for index, digit in enumerate(main_digits_list):
        digits_sum += digit * (10 - index)

    result = 11 - (digits_sum % 11)

    if result == 11:
        return '0'
    elif result == 10:
        return 'X'
    else:
        return str(result)

def calculate_check_digit_13(main_digits_list):
    digits_sum = 0
    # Multiplica os 12 primeiros dígitos por 1 e 3 alternadamente
    for index, digit in enumerate(main_digits_list):
        if index % 2 == 0:
            digits_sum += digit * 1
        else:
            digits_sum += digit * 3

    result = 10 - (digits_sum % 10)

    if result == 10:
        return '0'
    else:
        return str(result)

def main():
    user_input = input('Enter ISBN and length: ')
    if "," not in user_input:
        print("Enter comma-separated values.")
        return

    values = user_input.split(',')
    isbn = values[0].strip()
    length_str = values[1].strip()

    if not length_str.isdigit():
        print("Length must be a number.")
        return

    length = int(length_str)

    # Valida se o ISBN contém apenas números (ou termina com X se for tamanho 10)
    if length == 10 and isbn[-1].upper() == 'X':
        if not isbn[0:9].isdigit():
            print("Invalid character was found.")
            return
    elif not isbn.isdigit():
        print("Invalid character was found.")
        return

    if length == 10 or length == 13:
        validate_isbn(isbn, length)
    else:
        print('Length should be 10 or 13.')

# Remova o comentário abaixo para executar o programa diretamente
main()
```
